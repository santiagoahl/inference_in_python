# Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as st
from scipy.stats import f_oneway, ttest_ind
```

# Data import

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
iris = pd.read_csv(url, names=names)
```

```python
iris.head()
```

# Hypothesis Tests

## 1. $t$-Student
image.png

```python
se_length = iris['sepal-length'].std()/np.sqrt(10)
se_width = iris['sepal-length'].std()/np.sqrt(10)
```

image.png

```python
sed = np.sqrt((se_length)**2+(se_width)**2)
```

Formula of $t$-Student 

image.png

```python
t_stat = (iris['sepal-length'].mean()-iris['sepal-width'].mean()) / sed
t_stat
```

```python
p = ttest_ind(iris['sepal-length'], iris['sepal-width'])
p
```

We conclude that the hypothesis $H_0$:=both variables have similar distribution. This conclusion is caused of statistic is greater than the $p$ value 

## 2. Pearson coefficient

```python
corr = iris['sepal-length'].corr(iris['sepal-width'])
corr
```

```python
pearson = st.pearsonr(iris['sepal-length'], iris['sepal-width'])
pearson
```

## 2. ANOVA

```python
anova = f_oneway(iris['sepal-length'], iris['sepal-width'])
anova
```

We conclude that the hypothesis $H_0$:=both variables have similar distribution. This conclusion is caused of statistic is greater than the $p$ value
