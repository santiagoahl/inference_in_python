## 2) Systematic sampling

```python
def sys_sampling(data, step):
  ind = np.arange(0, len(data), step=step)
  sample = data.iloc[ind]
  return sample
```

```python
systematic_sample = sys_sampling(econdata, 3)
systematic_sample
```

## 3) Stratified sampling

```python
econdata['estratificado'] = econdata['delegacion']+', '+econdata['tipo']
(econdata['estratificado'].value_counts() / len(econdata)).sort_values(ascending=False)
```

Data shows the following proportion

1. Cuautémoc hotels: 64.3%
2. Cuautémoc museums: 15.6%
3. Venustiano Carranza: 7.8%
4. Cuauhtémoc: 7.3%
5. Venustiano Carranza: 4.7%


```python
def strat_sampling(data, strat_col_names, strat_values, strat_prop, random_state=None):
  df_strat = pd.DataFrame(columns=data.columns)
  pos=-1
  for i in range(len(strat_values)):
    pos+=1
    if pos == len(strat_values)-1:
      ration_len = len(data)-len(df_strat)
    else:
      ratio_len = int(len(data)*strat_prop[i])
    
    df_filtered = data[data[strat_col_names]==strat_values[i]]
    df_temp = df_filtered.sample(replace=True, n=ratio_len, random_state=random_state)

    df_strat = pd.concat([df_strat, df_temp])
  return df_strat
```

```python
strat_values = ['Cuautémoc, Hotel' , 'Cuautémoc, Museo', 'Venustiano Carranza, Hotel', 'Cuauhtémoc, Mercado', 'Venustiano Carranza, Mercado']
strat_prop = [0.643478, 0.156522, 0.078261, 0.073913, 0.047826]
df_strat = strat_sampling(econdata, 'estratificado', strat_values, strat_prop, random_state=42)
```
