# Bootstrapping

We'll use this re-sampling technique in order to compute the mean of the whole population

```python
def bootstrapp(data, sample_size, sub_sample_size):
  sample_means = []
  data_list = data.tolist()
  for i in range(sample_size // sub_sample_size):
    sample = random.sample(data_list, sub_sample_size)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)
  return np.mean(sample_means)
```
