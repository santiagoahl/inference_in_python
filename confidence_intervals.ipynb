# Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
```

# Data generation

```python
np.random.seed(20)
days = np.random.randint(0, 10, 10)
costs = np.random.normal(0, 1, 10)    
var = days + costs # Vacation days
```

```python
days
```

# Data viz

Results visualization with 80% of confidence

```python
graph = sns.regplot(days, var, ci=100)
plt.show()
```

```python
tuple = list(zip(days, var))
tuple
```

```python
table = pd.DataFrame(tuple, columns = ['Vacation days', 'Money spent'])
```

```python
table
```

## 95% of confidence

```python
st.t.interval(alpha=0.95, df=len(table)-1, loc=np.mean(table), scale=st.sem(table))
```
